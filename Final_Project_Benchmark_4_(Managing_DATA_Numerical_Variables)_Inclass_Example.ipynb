```python
#Create new variable 'current_weight', remove refuse, skip, and idk, always check the max
df['current_weight'] = df['H4GH6']
df['current_weight'] = pd.to_numeric( df['current_weight'] , errors ='coerce')
df['current_weight'] = df['current_weight'].loc[ df['current_weight'] != ' ' ]
df['current_weight'] = df['current_weight'].loc[ df['current_weight'] < 996 ]
df['current_weight'].describe()

```

# **Special Numerical Variables**

Special numerical variable.These variable include **months and years**, **feet and inches**, also **wake and bedtimes**.

```python
#Create new variable
df['combat_years'] = df['H4MI12Y']
df['combat_years'] = pd.to_numeric( df['combat_years'] , errors ='coerce')
df['combat_years'] = df['combat_years'].loc[ df['combat_years'] != ' ' ]
df['combat_years'] = df['combat_years'].loc[ df['combat_years'] < 97 ]
df['combat_years'].describe()
```

```python
#Create combat months
df['combat_months'] = df['H4MI12M']
df['combat_years'] = pd.to_numeric( df['combat_months'] , errors ='coerce')
df['combat_months'] = df['combat_months'].loc[ df['combat_months'] != ' ' ]
df['combat_months'] = df['combat_months'].loc[ df['combat_months'] < 97 ]
df['combat_months'].describe()
```

```python
#Now create a formula for a new variable combining months and years (Each year is 12 months)
df['comabt_time'] = df['combat_years'] * 12 + df['combat_months']
df['comabt_time'].describe()
```
