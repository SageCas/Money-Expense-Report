```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
link = 'https://raw.githubusercontent.com/kmath22/Math-200/refs/heads/main/Coin%20Circulation%20%20Research%201.0%20-%20Form%20Responses%201.csv'

df = pd.read_csv(link)

df
```

```python
# prompt: remove Age values greater than 200

# Assuming 'Age' is the column name containing the age values
df = df[df['Age'] <= 200]
```

```python
df.describe().round(2)
```

```python
fig,axes = plt.subplots(3,1, figsize =(8,7))
ax1 = sns.histplot( ax = axes[0], x ='Age', data =df,)
ax2 = sns.boxplot( ax = axes[1], x = 'Age' , data =df)
ax3 = sns.stripplot(ax = axes[2], x ='Age', data =df, size=5)
ax4 = sns.pointplot( ax = axes[2], x='Age', data = df, capsize=.5, errorbar=('ci' , 95), color ='r',)
ax1.axvline(x = 30, color = 'r', label = 'axvline - full height', linestyle = '--')
ax2.axvline(x = 30, color = 'r', label = 'axvline - full height', linestyle = '--')
ax3.axvline(x = 30, color = 'g', label = 'axvline - full height', linestyle = '--')
plt.show()
```
