```python
import pandas as pd
```

```python
#New Dataset with Demographics
url='https://drive.google.com/file/d/1J1lYAGJTivdEYYFoVK7m7tShm4kNhSgh/view?usp=sharing'
path= 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
df = pd.read_csv(path)
df.head(5)
```

```python
#Remove all blank responses
df= df.dropna()
```

```python
#frequency table for H4TO3
df['H4TO3'].value_counts()
```

```python
#Create new variable 'smoke_reg'
df['smoke_reg'] = df['H4TO3']
df['smoke_reg'].value_counts()
```

```python
#Remove response of refused and I don't know (6 and 8)
df = df[df.smoke_reg != 6]
df = df[df.smoke_reg != 8]
df['smoke_reg'].value_counts()
```

```python
#Format numbers to names
df['smoke_reg'].replace( [0,1,7] , ['NO','YES','NO'] ,  inplace= True )
df['smoke_reg'].value_counts()
```
